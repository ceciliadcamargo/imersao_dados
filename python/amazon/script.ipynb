# Importar pacotes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Importar base de dados

```python
dados = pd.read_csv("/content/Dados_Imersão - dados.csv")
avaliacoes = pd.read_csv("/content/Avaliacoes - avaliacoes.csv")
```

# Limpeza dos dados

```python
# Coluna dos preços
dados['actual_price'] = dados['actual_price'].str.replace('₹', '').str.replace('.', '').str.replace(',', '.')
dados['actual_price'] = pd.to_numeric(dados['actual_price'])
dados.info()
```

```python
# Coluna dos preços em real
dados['actual_price_real'] = dados['actual_price_real'].str.replace('R\$', '').str.replace('.', '').str.replace(',', '.').astype(float)
dados.info()
```

```python
# Coluna dos valores dos produtos
dados['product_price'] = dados['product_price'].str.replace('₹', '').str.replace('.', '').str.replace(',', '.')
dados['product_price'] = pd.to_numeric(dados['product_price'])
dados.info()
```

```python
# Coluna das notas
dados['rating'] = dados['rating'].str.replace(',', '.')
dados['rating'] = pd.to_numeric(dados['rating'])
dados.info()
```

```python
dados.describe()
```

```python
# SubstituIR os valores NaN por zero (0)
dados = dados.fillna(0)
```

# Juntando as bases de dados

```python
avaliacoes.head()
```

```python
# Adicionando os links das imagens dos produtos do data frame de avaliações ao data frame de dados
dados_avaliacao = pd.merge(dados, avaliacoes[['product_id', 'img_link']], on = 'product_id', how ='left')
dados_avaliacao.head()
```

# Gráficos

```python
# Selecionando apenas as notas acima de 4.5
ratings_above_4_5 = dados[dados["rating"] > 4.5]
rating_counts = ratings_above_4_5["rating"].value_counts().sort_index()

# Plotando o gráfico de barras
plt.figure(figsize = (10,6))
review_plot = sns.barplot(x = rating_counts.index, y = rating_counts.values, palette = "pastel", width = 0.5)
sns.despine()
review_plot.set_title("Contagem de Avaliações acima de 4.5", fontsize=16)
review_plot.set_xlabel("Avaliação", fontsize=14)
review_plot.set_ylabel("Contagem", fontsize=14)
review_plot.tick_params(axis="x", labelsize=12)
review_plot.tick_params(axis="y", labelsize=12)
plt.show()
```

```python
# Criando faixas de valores
faixa1 = dados[dados["actual_price"] <= 50]
faixa2 = dados[(dados["actual_price"] > 50) & (dados["actual_price"] <= 500)]
faixa3 = dados[dados["actual_price"] > 500]

# Contagem de produtos em cada faixa de valor
contagem_faixa1 = len(faixa1)
contagem_faixa2 = len(faixa2)
contagem_faixa3 = len(faixa3)

# Preparando os dados para o gráfico
contagens = [contagem_faixa1, contagem_faixa2, contagem_faixa3]
faixas = ["Até 50", "Entre 50 e 500", "Acima de 500"]
explode = (0.1, 0, 0)

# Plotando o gráfico de barras
plt.figure(figsize = (10,6))
count_plot = sns.barplot(x = faixas, y = contagens, palette = "Blues", width = 0.5)
sns.despine()
count_plot.set_title("Distribuição de produtos por faixas de valor", fontsize=16)
count_plot.set_xlabel("Valores", fontsize=14)
count_plot.set_ylabel("Produtos", fontsize=14)
count_plot.tick_params(axis="x", labelsize=12)
count_plot.tick_params(axis="y", labelsize=12)
plt.show()
```

```python
# Plotando gráfico de dispersão interativo
plt.figure(figsize = (10,6))
rating_plot = px.scatter(dados, x='rating', y='rating_count', title='Notas dos produtos')

rating_plot.update_layout(
    title_font_size=30,  # Tamanho do título
    xaxis_title='Notas',  # Título do eixo X
    yaxis_title='Votos',  # Título do eixo Y
    xaxis=dict(tickfont=dict(size=14)),  # Tamanho da fonte dos rótulos do eixo X
    yaxis=dict(tickfont=dict(size=14)),   # Tamanho da fonte dos rótulos do eixo Y
    xaxis_showgrid=False,  # Mostrar linhas de grade no eixo X
    yaxis_showgrid=False,   # Mostrar linhas de grade no eixo Y
    plot_bgcolor='rgba(0,0,0,0)'
)

rating_plot.update_traces(
    marker=dict(color='#5B9A8B'),
    hovertemplate='Nota: %{x}<br>Votos: %{y}')

rating_plot.write_html('rating_plot.html')
```

```python
# Agrupando os dados pelas categorias e fazendo a contagem
contagem_categorias = dados.groupby('category_geral').size().reset_index(name='contagem')

# Fazendo a média
media_preco_categorias = dados.groupby('category_geral')['actual_price_real'].mean().reset_index(name = 'média')

# Juntando os data frames
contagem_preco_categorias = pd.merge(contagem_categorias, round(media_preco_categorias, 2), on='category_geral')
contagem_preco_categorias
```

```python
category_plot = px.bar(contagem_preco_categorias, x='contagem', y='category_geral',
                       title='Categorias dos Produtos', hover_data=['contagem'])

category_plot.update_layout(
    title_font_size=30,  # Tamanho do título
    xaxis_title='Quantidade',  # Título do eixo X
    yaxis_title='',  # Título do eixo Y
    xaxis=dict(tickfont=dict(size=14)),  # Tamanho da fonte dos rótulos do eixo X
    yaxis=dict(tickfont=dict(size=14)),   # Tamanho da fonte dos rótulos do eixo Y
    xaxis_showgrid=False,  # Mostrar linhas de grade no eixo X
    yaxis_showgrid=False,   # Mostrar linhas de grade no eixo Y
    plot_bgcolor='rgba(0,0,0,0)'
)

category_plot.update_traces(
    marker_color='#F0B86E',
    hovertemplate='Quantidade: %{x}')

category_plot.write_html('category_plot.html')
```

```python
price_plot = px.bar(contagem_preco_categorias, x='média', y='category_geral', title='Preço médio das categorias')

price_plot.update_layout(
    title_font_size=30,  # Tamanho do título
    xaxis_title='Preços',  # Título do eixo X
    yaxis_title='',  # Título do eixo Y
    xaxis=dict(tickfont=dict(size=14)),  # Tamanho da fonte dos rótulos do eixo X
    yaxis=dict(tickfont=dict(size=14)),   # Tamanho da fonte dos rótulos do eixo Y
    xaxis_showgrid=False,  # Mostrar linhas de grade no eixo X
    yaxis_showgrid=False,   # Mostrar linhas de grade no eixo Y
    plot_bgcolor='rgba(0,0,0,0)'
)

price_plot.update_traces(
    marker_color='#836096',
    hovertemplate='Preço médio: R$ %{x}')

price_plot.write_html('price_plot.html')
```
